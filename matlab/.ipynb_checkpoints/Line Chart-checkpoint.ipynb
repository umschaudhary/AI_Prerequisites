```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,10)
# start point,end point , no of points in between
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel('TIme')
plt.ylabel('Some function of time')
plt.title("My cool chart")
```
